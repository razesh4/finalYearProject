```python
import pandas as pd
import tensorflow
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten

from tensorflow.keras import layers
```

```python
df=pd.read_csv('/kaggle/input/rajesh42ka1/1000_datapoint_duplicated_data_for_ann.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train,x_test,y_train,y_test=train_test_split(df.drop('name',axis=1),df['name'],test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train_encoder=encoder.fit_transform(y_train)
y_test_encoder=encoder.fit_transform(y_test)
```

```python
model = Sequential()
model.add(Flatten(input_shape=(x_train.shape[1],)))
model.add(Dense(1000,activation='relu')),
model.add(Dense(1000,activation='relu')),
model.add(Dense(1000,activation='relu')),

model.add(Dense(1000,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(x_train_scaled,y_train_encoder,epochs=100,validation_split=0.2,batch_size=1,verbose=1)
```

```python
model.save('/kaggle/working/')
```
